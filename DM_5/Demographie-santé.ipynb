# Nettoyage des données

Nettoyage des dataframes : suppression de certaines lignes, renommage de colonnes, modifications de certains string

```python
specialistes = specialistes[specialistes["Spécialistes"].str.contains("TOTAL")==False]
specialistes = specialistes[specialistes["REGION INSEE"].str.contains("TOTAL")==False]
specialistes = specialistes.dropna(subset=["DEPASSEMENTS /APE (Euros)"])
specialistes = specialistes.rename(columns={"Spécialistes":"Spécialité","REGION INSEE":"Régions"})
specialistes["Régions"] = specialistes["Régions"].str.replace("\d\d- ","")
specialistes["Spécialité"] = specialistes["Spécialité"].str.replace("\d\d- ","")

generalistes = generalistes[generalistes["Généralistes et compétences MEP"].str.contains("TOTAL")==False]
generalistes = generalistes[generalistes["REGION INSEE"].str.contains("TOTAL")==False]
generalistes = generalistes.dropna(subset=["DEPASSEMENTS /APE (Euros)"])
generalistes = generalistes.rename(columns={"Généralistes et compétences MEP":"Spécialité","REGION INSEE":"Régions"}) 
generalistes["Régions"] = generalistes["Régions"].str.replace("\d\d- ","")
generalistes["Spécialité"] = generalistes["Spécialité"].str.replace("\d\d- ","")

dentistes = dentistes[dentistes["Chirurgiens-dentistes et ODF"].str.contains("TOTAL")==False]
dentistes = dentistes[dentistes["REGION INSEE"].str.contains("TOTAL")==False]
dentistes = dentistes.dropna(subset=["DEPASSEMENTS /APE (Euros)"])
dentistes = dentistes.rename(columns={"Chirurgiens-dentistes et ODF":"Spécialité","REGION INSEE":"Régions"})
dentistes["Régions"] = dentistes["Régions"].str.replace("\d\d- ","")
dentistes["Spécialité"] = dentistes["Spécialité"].str.replace("\d\d- ","")

sages_femmes = sages_femmes[sages_femmes["Sages-femmes"].str.contains("TOTAL")==False]
sages_femmes = sages_femmes[sages_femmes["REGION INSEE"].str.contains("TOTAL")==False]
sages_femmes = sages_femmes.dropna(subset=["DEPASSEMENTS /APE (Euros)"])
sages_femmes = sages_femmes.rename(columns={"Sages-femmes":"Spécialité","REGION INSEE":"Régions"})
sages_femmes["Régions"] = sages_femmes["Régions"].str.replace("\d\d- ","")
sages_femmes["Spécialité"] = sages_femmes["Spécialité"].str.replace("\d\d- ","")

auxiliaires = auxiliaires[auxiliaires["Auxiliaires médicaux"].str.contains("TOTAL")==False]
auxiliaires = auxiliaires[auxiliaires["REGION INSEE"].str.contains("TOTAL")==False]
auxiliaires = auxiliaires.dropna(subset=["DEPASSEMENTS /APE (Euros)"])
auxiliaires = auxiliaires.rename(columns={"Auxiliaires médicaux":"Spécialité","REGION INSEE":"Régions"})
auxiliaires["Régions"] = auxiliaires["Régions"].str.replace("\d\d- ","")
auxiliaires["Spécialité"] = auxiliaires["Spécialité"].str.replace("\d\d- ","")
```

```python
population = population[population["Régions"].str.contains("métropo")==False]
population = population[population["Régions"].str.contains("DOM")==False]
```

Création d'un dataframe regroupant toutes les spécialités et nettoyage de la colonne Régions pour que les valeurs coincident entre les dataframes medecins et population

```python
medecins = pd.concat([specialistes[["Spécialité","Régions","NOMBRE D\'APE","DEPASSEMENTS /APE (Euros)"]],\
                      generalistes[["Spécialité","Régions","NOMBRE D\'APE","DEPASSEMENTS /APE (Euros)"]],\
                     dentistes[["Spécialité","Régions","NOMBRE D\'APE","DEPASSEMENTS /APE (Euros)"]],\
                     sages_femmes[["Spécialité","Régions","NOMBRE D\'APE","DEPASSEMENTS /APE (Euros)"]],\
                     auxiliaires[["Spécialité","Régions","NOMBRE D\'APE","DEPASSEMENTS /APE (Euros)"]]])
```

```python
medecins["Régions"] = medecins["Régions"].str.replace(" et ","-")
medecins["Régions"] = medecins["Régions"].str.replace("Centre-Val-de-Loire","Centre-Val de Loire")
medecins["Régions"] = medecins["Régions"].str.replace("Grand-Est","Grand Est")
medecins["Régions"] = medecins["Régions"].str.replace("Ile-de-France","Île-de-France")
medecins["Régions"] = medecins["Régions"].str.replace("Nouvelle Aquitaine","Nouvelle-Aquitaine")
medecins["Régions"] = medecins["Régions"].str.replace("Pays-de-Loire","Pays de la Loire")
medecins["Régions"] = medecins["Régions"].str.replace("Réunion","La Réunion")
population["Régions"] = population["Régions"].str.replace("Guadeloupe ","Guadeloupe")
population["Régions"] = population["Régions"].str.replace("Martinique ","Martinique")
```

# Modeles

Création de méthodes pour faire une régression linéaire entre le dépassement d'honoraire d'une spécialité et le nombre de practiciens de cette spécialité par région 

```python
def depassement_honoraires(speciality):
    
    lr = LinearRegression()
    lr.fit(medecins[medecins["Spécialité"]==speciality]["NOMBRE D\'APE"][:, None], medecins[medecins["Spécialité"]==speciality]["DEPASSEMENTS /APE (Euros)"])

    y = lr.predict(medecins[medecins["Spécialité"]==speciality]["NOMBRE D\'APE"][:, None])

   
    return y,lr.score(medecins[medecins["Spécialité"]==speciality]["NOMBRE D\'APE"][:, None], medecins[medecins["Spécialité"]==speciality]["DEPASSEMENTS /APE (Euros)"])
```

```python
def plot_regression(speciality) :
    plt.scatter(medecins[medecins["Spécialité"]==speciality]["NOMBRE D\'APE"], medecins[medecins["Spécialité"]==speciality]["DEPASSEMENTS /APE (Euros)"])
    plt.plot(medecins[medecins["Spécialité"]==speciality]["NOMBRE D\'APE"],depassement_honoraires(speciality)[0],color="red")
    plt.title(speciality)
    plt.xlabel("Nombre de practiciens par région")
    plt.ylabel("Dépassement d'honoraires en euros")
    plt.show()
```

## Corrélations entre la densité de médecins par spécialité et par région et la pratique du dépassement d'honoraires

Top 10 des corrélations 

```python
for specialite in specialites:
    if depassement_honoraires(specialite)[1] > 0.75 and \
    len(medecins[medecins["Spécialité"]==specialite]["NOMBRE D\'APE"].unique())>3:
        print(specialite)
```

## Corrélation entre le nombre d'infirmiers et le nombre de personnes de plus de 75 ans par région

```python
infirmiers = pd.merge(medecins[(medecins["Spécialité"]=="Infirmiers")],\
                      population[["Régions","75 ans et plus"]],on="Régions")
```

```python
lr = LinearRegression()
lr.fit(infirmiers["75 ans et plus"][:, None],infirmiers["NOMBRE D\'APE"])

y = lr.predict(infirmiers["75 ans et plus"][:, None])

lr.score(infirmiers["75 ans et plus"][:, None],infirmiers["NOMBRE D\'APE"])
```

```python
plt.scatter(infirmiers["75 ans et plus"],infirmiers["NOMBRE D\'APE"])
plt.plot(infirmiers["75 ans et plus"],y,color='red')
plt.show()
```
